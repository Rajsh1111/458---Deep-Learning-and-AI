# Raw Python Code
Source: https://towardsdatascience.com/perceptrons-logical-functions-and-the-xor-problem-37ca5025790a

Defining necessary functions:

```python
def unit_step(v):
    #Heavyside Step function. v must be a scalar
    if v >= 0:
        return 1
    else:
        return 0

def perceptron(x, w, b):
    #Function implemented by a perceptron with weight vector w and bias b
    v = np.dot(w, x) + b
    y = unit_step(v)
    return y

def NOT_percep(x):
    w = -1
    b = 0.5
    return perceptron(x, w, b)

def AND_percep(x):
    w = np.array([1, 1])
    b = -1.5
    return perceptron(x, w, b)

def OR_percep(x):
    w = np.array([1, 1])
    b = -0.5
    return perceptron(x, w, b)

def XOR_percep(x):
    gate_1 = AND_percep(x)
    gate_2 = NOT_percep(gate_1)
    gate_3 = OR_percep(x)
    new_x = np.array([gate_2, gate_3])
    output = AND_percep(new_x)
    return output
```

Defining Test arrays:

```python
example1 = np.array([1, 1])
example2 = np.array([1, 0])
example3 = np.array([0, 1])
example4 = np.array([0, 0])
```

Testing:

```python
print("AND({}, {}) = {}".format(1, 1, AND_percep(example1)))
print("AND({}, {}) = {}".format(1, 0, AND_percep(example2)))
print("AND({}, {}) = {}".format(0, 1, AND_percep(example3)))
print("AND({}, {}) = {}".format(0, 0, AND_percep(example4)))
print("-------------")
print("OR({}, {}) = {}".format(1, 1, OR_percep(example1)))
print("OR({}, {}) = {}".format(1, 0, OR_percep(example2)))
print("OR({}, {}) = {}".format(0, 1, OR_percep(example3)))
print("OR({}, {}) = {}".format(0, 0, OR_percep(example4)))
print("-------------")
print("XOR({}, {}) = {}".format(1, 1, XOR_percep(example1)))
print("XOR({}, {}) = {}".format(1, 0, XOR_percep(example2)))
print("XOR({}, {}) = {}".format(0, 1, XOR_percep(example3)))
print("XOR({}, {}) = {}".format(0, 0, XOR_percep(example4)))
```

# Using TensorFlow
Source: https://github.com/marjoleinspronk/XORproblem/blob/master/XOR_TensorFlow2.0.ipynb

Creating data sets:

```python
x_train, y_train = (tf.constant([[0,0],[0,1],[1,0],[1,1]], "float32"), tf.constant([[0],[1],[1],[0]], "float32"))
```

Creating Neural Network model:

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(2,)),
  tf.keras.layers.Dense(2, activation=tf.nn.relu),   # hidden layer
  tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)   # output layer
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

Fitting model:

```python
model.fit(x_train, y_train, epochs=1000)
```

Results:

```python
predictions = model.predict(x_train).round()
for item in range(4):
    print(x_train[item], " = ", predictions[item])
#print(x_train[0])
#print(model.predict(x_train).round())  
```
